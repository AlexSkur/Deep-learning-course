### **1.   ReLU activation function**


> $f(x) = max(0, x) = \frac{x+|x|}{2} = \begin{cases}
  x, & x > 0\\
  0, & x \leq 0
\end{cases}$





```python
def relu(x: np.ndarray) -> np.ndarray:
    return np.abs(x * (x > 0.0))
    # return np.maximum(0, x)
```

```python
relu_func = relu(x)
print(relu_func)
```

```python
plt.plot(x, relu_func, label='activation')
plt.title("ReLU")
plt.xlabel('input')
plt.ylabel('output')
plt.grid()
plt.show()
```

### Derivative of ReLU function


> $f'(x) =
  \begin{cases}
    1, & x > 0\\
    0, & x \leq 0
  \end{cases}
$


```python
def derivative_relu(x: np.ndarray) -> np.ndarray:
  return np.where(x > 0, 1.0, 0.0)
```

```python
deriv_relu = derivative_relu(x)
```

```python
plt.plot(x, deriv_relu, c='g')
plt.title("Derivative of ReLU")
plt.xlabel("input")
plt.ylabel("output")
plt.grid()
plt.show()
```

```python
plt.plot(x, relu_func)
plt.plot(x, deriv_relu, c='g')
plt.title("RELU and derivative")
plt.xlabel("input")
plt.ylabel("output")
plt.yticks(np.arange(0, 11))
plt.grid()
plt.show()
```

### **2. ReLU6 activation function**

    0, & x \leq 0\\
    x, & x > 0\\
    6, & x \geq 6

### **3. ELU (Exponential Linear Unit) activation function**
> $f(x) = \begin{cases}
  x, & x\geq 0\\
  \alpha \cdot (e^{x} - 1), & x < 0
\end{cases}$

```python
def elu(x: np.ndarray, alpha: float) -> np.ndarray:
  res = np.where(x >= 0, x, alpha * (np.exp(x) - 1))
  return res
```

```python
alpha = 0.3
elu_func = elu(x, alpha=alpha)
print(elu_func)
```

```python
plt.plot(x, elu_func, label='activation')
plt.title("ELU")
plt.xlabel('input')
plt.ylabel('output')
plt.legend()
plt.grid()
plt.show()
```

### Derivative of ELU
> $ELU'(x) = \begin{cases}
  1, & x \geq 0\\
  \alpha \cdot e^{x}, & x < 0
\end{cases} = \begin{cases}
  1, & x \geq 0\\
  ELU(x) + alpha, & x < 0
\end{cases}$

```python
def derivative_elu(x: np.ndarray, alpha: float) -> np.ndarray:
  return np.where(x >= 0, 1.0, elu(x, alpha) + alpha)
```

```python
# alpha = 0.3
deriv_elu = derivative_elu(x, alpha)
```

```python
plt.plot(x, deriv_elu)
plt.title("Derivative of ELU")
plt.xlabel("input")
plt.ylabel("output")
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, elu_func, label='activation')
plt.plot(x, deriv_elu, c='g', ls='--', label='derivative')
plt.title("ELU and derivative")
plt.xlabel('input')
plt.ylabel('output')
plt.yticks(np.arange(-1, 11))
plt.legend()
plt.grid()
plt.show()
```

For correct display the graph of the derivative of the ELU function:

```python
def derivative_elu_2(x: np.ndarray, alpha: float):
  res1 = np.zeros(x[x >= 0].shape)
  res2 = x[x < 0]
  res1[:] = 1.0
  res2[:] = elu(res2, alpha) + alpha
  return (res1, res2)
```

```python
# alpha = 0.3
deriv_elu = derivative_elu_2(x, alpha=alpha)
deriv_elu1 = deriv_elu[0]
deriv_elu2 = deriv_elu[1]
print(deriv_elu1.shape)
print("\n")
print(deriv_elu2.shape)
```

### **4. SiLU (Sigmoid Linear Units) activation function**
> **$SiLU(x) = x \cdot \sigma(x)$**
>>>  $\sigma(x) = \frac{1}{1+e^{-x}}$

```python
def sigmoid(x: np.ndarray) -> np.ndarray:
  return (1.0 / (1.0 + np.exp(-x)))
```

```python
def silu(x: np.ndarray) -> np.ndarray:
  return x * sigmoid(x)
```

```python
silu_func = silu(x)
print(silu_func)
```

```python
plt.plot(x, silu_func, label='SiLU')
plt.title("SiLU")
plt.xlabel("input")
plt.ylabel("output")
plt.legend()
plt.grid()
plt.show()
```

### Derivative of sigmoid
> $\sigma'(x) = \displaystyle (\frac{e^{x}}{1+e^{x}})' = \frac{e^x \cdot (1+e^x) - e^x \cdot e^x}{(1+e^x)^2} = e^x \cdot (\frac{1}{1+e^x} - \frac{e^x}{(1+e^x)^2}) = \frac{e^x}{1+e^x} - \frac{e^{2x}}{(1+e^x)^2} =
\sigma(x) - \sigma^2(x) = \sigma(x)(1 - \sigma(x))\\[1cm]$

### Derivative of SiLU
> $SiLU'(x) = (x \cdot \sigma(x))' = \sigma(x) + x \cdot \sigma'(x) = \sigma(x) + x \cdot \sigma(x)(1 - \sigma(x)) = \sigma(x)(1 + x \cdot (1 - \sigma(x))) = \sigma(x) + SiLU(x) \cdot (1 - \sigma(x)) = \mathbf{\sigma(x)(1 + x - SiLU(x))}$

```python
def derivative_silu(x: np.ndarray) -> np.ndarray:
  return sigmoid(x) * (1.0 + x - silu(x))
```

```python
deriv_silu = derivative_silu(x)
print(deriv_silu)
```

```python
plt.plot(x, deriv_silu, c='g', ls='--', label='derivative')
plt.title("Derivative of SiLU")
plt.xlabel("input")
plt.ylabel("output")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(x, silu_func, label='SiLU')
plt.plot(x, deriv_silu, c='g', ls='--', label='derivative')
plt.title("SiLU and derivative")
plt.xlabel('input')
plt.ylabel('output')
plt.legend()
plt.grid()
plt.show()
```

### **6. Mish activation function**
> $f(x) = x \cdot tanh(softplus(x))$

```python
def mish(x: np.ndarray) -> np.ndarray:
  return x * np.tanh(softplus(x))
```

```python
mish_func = mish(x)
print(mish_func)
```

```python
plt.plot(x, mish_func)
plt.title("Mish")
plt.xlabel('input')
plt.ylabel('output')
plt.grid()
plt.show()
```

### **Derivative of Mish**
> $mish'(x) = (x \cdot tanh(softplus(x)))' = \mathbf{tanh(softplus(x)) + x \cdot \displaystyle \frac{1}{cosh^2(softplus(x))} \cdot softplus'(x)}$

```python
def derivative_mish(x: np.ndarray) -> np.ndarray:
  softplus_val = softplus(x)
  res = np.tanh(softplus_val) + (x * derivative_softplus(x)) / (np.cosh(softplus_val) * np.cosh(softplus_val))
  return res
```

```python
deriv_mish = derivative_mish(x)
print(deriv_mish)
```

```python
plt.plot(x, deriv_mish, c='g', ls='--')
plt.title("Derivative of Mish")
plt.xlabel('input')
plt.ylabel('output')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.plot(x, mish_func, label='Mish')
plt.plot(x, deriv_mish, c='g', ls='--', label='derivative')
plt.title("Mish and derivative")
plt.xlabel('input')
plt.ylabel('output')
plt.yticks(np.arange(-1, 11))
plt.legend()
plt.grid()
plt.show()
```

### **7. Swish activation function**
> $Swish(x) = x \cdot \sigma(\beta \cdot x)$

```python
def swish(x: np.ndarray, beta: float) -> np.ndarray:
  return x * sigmoid(beta * x)
```

```python
beta = 0.1
swish_func = swish(x, beta)
print(swish_func)
```

```python
plt.plot(x, swish_func, c='b', label='beta = 0.1')
plt.plot(x, swish(x, 1.0), c='r', label='beta = 1.0')
plt.plot(x, swish(x, 10.0), c='y', label='beta=10.0')
plt.title(r'Swish with different $\beta$')
plt.xlabel('input')
plt.ylabel('output')
plt.legend()
plt.grid()
plt.show()
```

### **Derivative of Swish**
> $Swish'(x) = \sigma(\beta \cdot x) + x \cdot \sigma'(\beta \cdot x) \cdot \beta = \sigma(\beta \cdot x) + x \cdot \sigma(\beta \cdot x)(1 - \sigma(\beta \cdot x)) \cdot \beta = \\
\mathbf{\sigma(\beta \cdot x) \cdot (1 + x \cdot \beta \cdot (1 - \sigma(\beta \cdot x)))}$

```python
def derivative_swish(x: np.ndarray, beta: float) -> np.ndarray:
  beta_mul_x = beta * x
  sigmoid_beta_x = sigmoid(beta_mul_x)
  return sigmoid_beta_x * (1.0 + beta_mul_x * (1.0 - sigmoid_beta_x))
```

```python
deriv_swish = derivative_swish(x, beta=0.1)
print(deriv_swish)
```

```python
plt.plot(x, deriv_swish, c='b', label='beta = 0.1')
plt.plot(x, derivative_swish(x, beta=1.0), c='r', label='beta = 1.0')
plt.plot(x, derivative_swish(x, beta=10.0), c='y', label='beta=10.0')
plt.title(r'Derivative of Swish with different $\beta$')
plt.xlabel('input')
plt.ylabel('output')
plt.legend()
plt.grid()
plt.show()
```

```python
x1 = np.linspace(-1.5, 1.5, 200)
plt.plot(x1, derivative_swish(x, beta=0.1), c='b', label='beta = 0.1')
plt.plot(x1, derivative_swish(x, beta=1.0), c='r', label='beta = 1.0')
plt.plot(x1, derivative_swish(x, beta=10.0), c='y', label='beta=10.0')
plt.title(r'Derivative of Swish with different $\beta$')
plt.xlabel('input')
plt.ylabel('output')
plt.legend()
plt.grid()
plt.show()
```

\\[1cm]$

$Swish'(x) = \sigma(\beta \cdot x) + x \cdot \sigma'(\beta \cdot x) \cdot \beta = \sigma(\beta \cdot x) + x \cdot \sigma(\beta \cdot x)(1 - \sigma(\beta \cdot x)) \cdot \beta =\\ \beta \cdot x \cdot \sigma(\beta \cdot x) - \beta \cdot x \cdot \sigma^2(\beta \cdot x) + \sigma(\beta \cdot x) = \beta \cdot x \cdot \sigma(\beta \cdot x) + \sigma(\beta \cdot x) \cdot (1 - \beta \cdot x \cdot \sigma(\beta \cdot x)) =
\mathbf{\beta \cdot Swish(x) + \sigma(\beta \cdot x) \cdot (1 - \beta \cdot Swish(x))}$

```python
def derivative_swish_2(x: np.ndarray, beta: float) -> np.ndarray:
  # beta_x = beta * x
  # return sigmoid(beta_x) + beta * swish(x, beta) * (1 - sigmoid(beta_x))

  beta_swish = beta * swish(x, beta)
  return beta_swish + sigmoid(beta * x) * (1 - beta_swish)
```

### **1. BCE (Binary Cross-Entropy Loss)**
> $BCE = (-1) \cdot [y_i \cdot log(\hat{y_i}) + (1 - y_i) \cdot log(1 - \hat{y_i})]\\[0.5cm]$

where
> **$y$ - the true label, which is either 0 or 1,**

> **$y$ - the true label, which is either 0 or 1,**

> **$\hat{y}$ - predicted probability of the positive class (class 1). The predicted probability of the negative class (class 0) is simply $(1-\hat{y})$**


```python
def bce_loss(y_pred: np.ndarray, y_true: np.ndarray) -> np.float64:
  res = (-1) * (y_true * np.log(y_pred) + (1.0 - y_true) * np.log(1.0 - y_pred))
  return res
```

```python
y_pred = np.linspace(0.001, 1.0, 100, endpoint = False)
y_true = np.ones(y_pred.shape)
bce_y_1 = bce_loss(y_pred, y_true)
```

```python
plt.plot(y_pred, bce_y_1)
plt.title("BCE (y_true = 1)")
plt.xlabel('y_pred')
plt.ylabel('Loss')
plt.grid()
plt.show()
```

### **2. Kullback-Leibler divergence (KLDiv)**
> **KLDiv** = $y \cdot log(\frac{y}{\hat{y}}) = y \cdot (log(y) - log(\hat{y}))\\[0.5cm]$

where
> **$y$ - the true label, which is either 0 or 1,**

> **$\hat{y}$ - predicted probability of the positive class (class 1). The predicted probability of the negative class (class 0) is simply $(1-\hat{y})$**

```python
def kldiv_loss(y_pred: np.ndarray, y_true: np.ndarray) -> np.float64:
  return y_true * (np.log(y_true) - np.log(y_pred))
```

```python
y_pred = np.linspace(0.001, 1.0, 100, endpoint=False)
y_true = np.ones(y_pred.shape)

loss = kldiv_loss(y_pred, y_true);
```

```python
plt.plot(y_pred, loss)
plt.title("KLDiv (y_true = 1)")
plt.xlabel('y_pred')
plt.ylabel('KLDiv loss')
plt.grid()
plt.show()
```

### **3. Huber Loss**
> **Huber** =
$\begin{cases}
  \frac{1}{2} \cdot (\hat{y} - y)^2, & |\hat{y} - y| \leq \delta \\
  \delta \cdot (|\hat{y} - y| - \frac{1}{2} \delta), & otherwise
\end{cases} \\[0.5cm]$

where
> $y$ - is the true value or target value

> $\hat{y}$ - is the predicted value

> $\delta$ - is the delta parameter, which determines the threshold for switching between the quadratic and linear components of the loss function

```python
def huber_loss(y_pred: np.ndarray, y_true: np.ndarray, delta: float) -> np.float64:
  abs_diff = np.abs(y_pred - y_true)
  loss = np.where(abs_diff <= delta, 0.5 * abs_diff * abs_diff, delta * (abs_diff - 0.5 * delta))

  return loss
```

```python
y_true = np.linspace(-10, 10, 100)
y_pred = np.zeros(y_true.shape)
```

```python
deltas = [1, 3, 5, 0.5, 0.3]
colors = ['b', 'r', 'y', 'g', 'pink']

for i in range(len(deltas)):
  plt.plot(y_pred - y_true, huber_loss(y_pred, y_true, deltas[i]), c=colors[i], label=rf'$\delta$ = {deltas[i]}')

plt.title('Huber Loss')
plt.xlabel("y_pred - y_true")
plt.ylabel("Loss")
plt.legend()
plt.grid()
plt.show()
```

### **4. SmoothL1 Loss**
> **SmoothL1** =
$\begin{cases}
  \displaystyle \frac{0.5 \cdot (\hat{y} - y)^2}{\beta}, & |\hat{y} - y| < \beta \\
  |\hat{y} - y| - 0.5 \cdot \beta, & otherwise
\end{cases} \\[0.5cm]$

where
> $y$ - is the true value or target value

> $\hat{y}$ - is the predicted value

> $\beta$ - specifies the threshold at which to change between L1 and L2 loss. The value must be non-negative.

```python
def smoothL1_loss(y_pred: np.ndarray, y_true: np.ndarray, beta: float) -> np.float64:
  square_diff = (y_pred - y_true) * (y_pred - y_true)
  abs_diff = np.abs(y_pred - y_true)

  loss = np.where(abs_diff < beta, (0.5 * square_diff) / beta, abs_diff - 0.5 * beta)

  return loss
```

```python
y_true = np.linspace(-10, 10, 100)
y_pred = np.zeros(y_true.shape)
```

```python
betas = [1, 5, 100, 0.5, 0.01]
colors = ['b', 'r', 'y', 'g', 'pink']

for i in range(len(betas)):
  plt.plot(y_pred - y_true, smoothL1_loss(y_pred, y_true, betas[i]), c=colors[i], label=rf'$\beta$ = {betas[i]}')

plt.title('SmoothL1 Loss')
plt.xlabel("y_pred - y_true")
plt.ylabel("Loss")
plt.legend()
plt.grid()
plt.show()
```
